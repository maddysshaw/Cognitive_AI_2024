```python
# @title Calculate Distance Between Robots Over time

fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(5 * 4, 4))
fig.suptitle(f"Distance between Leader vs Follower for Turn Sharpness (20,15)")

mse_values_15 = []

for i, dL in enumerate([dL_15_take1, dL_15_take2, dL_15_take3, dL_15_take4, dL_15_take5]):
  df = [df_15_take1, df_15_take2, df_15_take3, df_15_take4, df_15_take5][i]
  # Compute Euclidean distance
  df["Distance"] = np.sqrt(((dL["X"]+90) - df["X"])**2 + (df["Y"] - dL["Y"])**2) - 40
  mse = np.mean((df["Distance"]-50)**2)
  mse_values_15.append(mse)
  print(f"Mean Squared Error (MSE) for repeat {i+1}: {mse:.4f}")
  axes[i].plot(df["Distance"], label="Distance", color="purple")
  axes[i].plot(df.index, [50]*len(df.index), color='r', linestyle='dashed', label="Ideal Distance (5cm)")
  axes[i].set_title(f"Repeat {i+1}")
  axes[i].set_xlabel("time")
  axes[i].set_ylabel("Distance")
  axes[i].set_ylim(0, 100)
  axes[i].legend()

plt.tight_layout()
plt.show()

fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(5 * 4, 4))
fig.suptitle(f"Distance between Leader vs Follower for Turn Sharpness (20,10)")

mse_values_10 = []

for i, dL in enumerate([dL_10_take1, dL_10_take2, dL_10_take3, dL_10_take4, dL_10_take5]):
  df = [df_10_take1, df_10_take2, df_10_take3, df_10_take4, df_10_take5][i]
  # Compute Euclidean distance
  df["Distance"] = np.sqrt(((dL["X"]+90) - df["X"])**2 + (df["Y"] - dL["Y"])**2) - 40
  mse = np.mean((df["Distance"]-50)**2)
  mse_values_10.append(mse)
  print(f"Mean Squared Error (MSE) for repeat {i+1}: {mse:.4f}")
  axes[i].plot(df["Distance"], label="Distance", color="purple")
  axes[i].plot(df.index, [50]*len(df.index), color='r', linestyle='dashed', label="Ideal Distance (5cm)")
  axes[i].set_title(f"Repeat {i+1}")
  axes[i].set_xlabel("time")
  axes[i].set_ylabel("Distance")
  axes[i].set_ylim(0, 100)
  axes[i].legend()

plt.tight_layout()
plt.show()

fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(5 * 4, 4))
fig.suptitle(f"Distance between Leader vs Follower for Turn Sharpness (20,5)")

mse_values_5 = []

for i, dL in enumerate([dL_5_take1, dL_5_take2, dL_5_take3, dL_5_take4, dL_5_take5]):
  df = [df_5_take1, df_5_take2, df_5_take3, df_5_take4, df_5_take5][i]
  # Compute Euclidean distance
  df["Distance"] = np.sqrt(((dL["X"]+90) - df["X"])**2 + (df["Y"] - dL["Y"])**2) - 40
  mse = np.mean((df["Distance"]-50)**2)
  mse_values_5.append(mse)
  print(f"Mean Squared Error (MSE) for repeat {i+1}: {mse:.4f}")
  axes[i].plot(df["Distance"], label="Distance", color="purple")
  axes[i].plot(df.index, [50]*len(df.index), color='r', linestyle='dashed', label="Ideal Distance (5cm)")
  axes[i].set_title(f"Repeat {i+1}")
  axes[i].set_xlabel("time")
  axes[i].set_ylabel("Distance")
  axes[i].set_ylim(0, 100)
  axes[i].legend()

plt.tight_layout()
plt.show()


fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(5 * 4, 4))
fig.suptitle(f"Distance between Leader vs Follower for Turn Sharpness (20,0)")

mse_values_0 = []

for i, dL in enumerate([dL_0_take1, dL_0_take2, dL_0_take3, dL_0_take4, dL_0_take5]):
  df = [df_0_take1, df_0_take2, df_0_take3, df_0_take4, df_0_take5][i]
  # Compute Euclidean distance
  df["Distance"] = np.sqrt(((dL["X"]+90) - df["X"])**2 + (df["Y"] - dL["Y"])**2) - 35
  mse = np.mean((df["Distance"]-50)**2)
  mse_values_0.append(mse)
  print(f"Mean Squared Error (MSE) for repeat {i+1}: {mse:.4f}")
  axes[i].plot(df["Distance"], label="Distance", color="purple")
  axes[i].plot(df.index, [50]*len(df.index), color='r', linestyle='dashed', label="Ideal Distance (5cm)")
  axes[i].set_title(f"Repeat {i+1}")
  axes[i].set_xlabel("time")
  axes[i].set_ylabel("Distance")
  axes[i].set_ylim(0, 100)
  axes[i].legend()

plt.tight_layout()
plt.show()
```

```python
# Compute mean distance over time for each sharpness value
# Get the minimum length among all DataFrames
min_len_15 = min(len(df) for df in [df_15_take1, df_15_take2, df_15_take3, df_15_take4, df_15_take5])
# Slice the 'Distance' columns to the minimum length before calculating the mean
mean_dist_15 = np.mean([df["Distance"][:min_len_15] for df in [df_15_take1, df_15_take2, df_15_take3, df_15_take4, df_15_take5]], axis=0)
std_dist_15 = np.std([df["Distance"][:min_len_15] for df in [df_15_take1, df_15_take2, df_15_take3, df_15_take4, df_15_take5]], axis=0)

min_len_10 = min(len(df) for df in [df_10_take1, df_10_take2, df_10_take3, df_10_take4, df_10_take5])
mean_dist_10 = np.mean([df["Distance"][:min_len_10] for df in [df_10_take1, df_10_take2, df_10_take3, df_10_take4, df_10_take5]], axis=0)
std_dist_10 = np.std([df["Distance"][:min_len_10] for df in [df_10_take1, df_10_take2, df_10_take3, df_10_take4, df_10_take5]], axis=0)

min_len_5 = min(len(df) for df in [df_5_take1, df_5_take2, df_5_take3, df_5_take4, df_5_take5])
mean_dist_5 = np.mean([df["Distance"][:min_len_5] for df in [df_5_take1, df_5_take2, df_5_take3, df_5_take4, df_5_take5]], axis=0)
std_dist_5 = np.std([df["Distance"][:min_len_5] for df in [df_5_take1, df_5_take2, df_5_take3, df_5_take4, df_5_take5]], axis=0)

min_len_0 = min(len(df) for df in [df_0_take1, df_0_take2, df_0_take3, df_0_take4, df_0_take5])
mean_dist_0 = np.mean([df["Distance"][:min_len_0] for df in [df_0_take1, df_0_take2, df_0_take3, df_0_take4, df_0_take5]], axis=0)
std_dist_0 = np.std([df["Distance"][:min_len_0] for df in [df_0_take1, df_0_take2, df_0_take3, df_0_take4, df_0_take5]], axis=0)

time_15 = np.arange(min_len_15)  # Ensure correct x-axis
time_10 = np.arange(min_len_10)
time_5 = np.arange(min_len_5)
time_0 = np.arange(min_len_0)

# Plot all mean distances
plt.figure(figsize=(6, 3.5))
plt.plot(mean_dist_15, label="Forward Bias 15", color="blue")
plt.fill_between(time_15, mean_dist_15 - std_dist_15, mean_dist_15 + std_dist_15,
                 color="blue", alpha=0.1)
plt.plot(mean_dist_10, label="Forward Bias 10", color="green")
plt.fill_between(time_10, mean_dist_10 - std_dist_10, mean_dist_10 + std_dist_10,
                 color="green", alpha=0.1)
plt.plot(mean_dist_5, label="Forward Bias 5", color="orange")
plt.fill_between(time_5, mean_dist_5 - std_dist_5, mean_dist_5 + std_dist_5,
                 color="orange", alpha=0.1)
plt.plot(mean_dist_0, label="Forward Bias 0", color="red")
plt.fill_between(time_0, mean_dist_0 - std_dist_0, mean_dist_0 + std_dist_0,
                 color="red", alpha=0.1)

# Desired distance line
plt.axhline(y=50, color='black', linestyle='dashed', label="Desired Distance")

# Labels and title
plt.xlabel("Time")
plt.ylabel("Mean Distance (mm)")
plt.title("Mean Distance Between Leader and Follower")
plt.legend()
plt.ylim(0,100)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
